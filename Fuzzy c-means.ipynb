```python
from sklearn import datasets
import numpy as np
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import skfuzzy
```

```python
iris = datasets.load_iris()
iris
```

```python
# iris.data.T = tranpor dados do dataset
# c = numero de clusters
# m = qual grupo cada cluster pertence. O valor 2 é o recomentado
# error = criterio de parada. O valor 0.005 é o recomentado
# maxiter = quantas vezes irá fazer a repetição
r = skfuzzy.cmeans(data = iris.data.T, c = 3, m=2, error = 0.005,
                  maxiter = 1000, init = None)
r
```

```python
previsoes_porcentagem = r[1]

#probabilidade de estar no grupo 0 
print(previsoes_porcentagem[0][0])
#probabilidade de estar no grupo 1 
print(previsoes_porcentagem[1][0])
#probabilidade de estar no grupo 2
print(previsoes_porcentagem[2][0])
```

```python
#pegar o grupo com amior porcentagem
# argmax(axis = 0) = maior valor por coluna
previsoes = previsoes_porcentagem.argmax(axis = 0)
previsoes
```

```python
resultados = confusion_matrix(iris.target, previsoes)
resultados
```
